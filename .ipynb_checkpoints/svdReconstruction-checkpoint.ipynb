```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(-10,10,1000)
h = -5*np.exp(-t**2) - np.random.rand(t.size)

s = np.exp(-(t-1)**2/0.01)+np.exp(-(t-8)**2/0.01)+np.exp(-(t+1)**2/0.01)+np.exp(-(t+8)**2/0.01)
hmat = np.zeros((h.size,h.size))
for i in range(h.size-1):
    for k in range(i+1):
        hmat[i][k]=h[i-k]
y = np.matmul(hmat,s)

P, D, Q = np.linalg.svd(hmat, full_matrices=False)
X = np.matmul(np.matmul(P, np.diag(D)), Q)

D_inv = 1/D
X_inv = np.matmul(np.matmul(np.linalg.inv(Q), np.diag(D_inv)), np.linalg.inv(P))
s_rec = np.matmul(X_inv,y)

fig = plt.figure()
plt.plot(t,s)
plt.plot(t,h)
#plt.plot(t,y)
plt.plot(t,s_rec)
plt.grid(True)
```
