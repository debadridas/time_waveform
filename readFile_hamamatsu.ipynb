```python
%matplotlib notebook

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_table('ave1/C1--HighPulse-in-100-out1700-an2100--00000.txt', sep="\t", names={'I','II'},
                   dtype = {'I': np.float128, 'II': np.float128}, usecols = [1,0], skiprows=5) 
df1 = pd.read_table('ave1/C1--LowPulse-in-100-out1700-an2100--00000.txt', sep="\t", names={'I','II'},
                   dtype = {'I': np.float128, 'II': np.float128}, usecols = [1,0], skiprows=5)
time = df['I'].tolist()
time = [ time[i]* 1.0e9 for i in range(len(time))]
ampl = df['II'].tolist()
time1 = df1['I'].tolist()
time1 = [ time1[i]* 1.0e9 for i in range(len(time1))]
ampl1 = df1['II'].tolist()

fig = plt.figure()
plt.plot(time,ampl)
plt.plot(time1,ampl1)
plt.grid('True')
```
